```python
import pandas as pd
from pprint import pprint
import json
from pandas.io.json import json_normalize
from sqlalchemy import create_engine
```

```python
US_df = pd.read_csv('ETL-Project/Resources/USvideos.csv')
CA_df = pd.read_csv('ETL-Project/Resources/CAvideos.csv')
GB_df = pd.read_csv('ETL-Project/Resources/GBvideos.csv')
```

```python
sub_json = []

us_json_file = ('ETL-Project/Resources/US_category_id.json')
us_json = pd.read_json(us_json_file)


for data in us_json['items']:
        sub_json.append([data['id'],data['snippet']['title']])

US_cat_df = pd.DataFrame.from_dict(sub_json)
US_cat_df.columns = ['category_id', 'category']
US_cat_df = US_cat_df.set_index('category_id')
US_cat_df
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/YouTube_Trends')
US_cat_df.to_sql(name = 'categories', con=engine, if_exists = 'append', index = True)
```
